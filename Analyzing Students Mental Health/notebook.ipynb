Does going to university in a different country affect your mental health? A Japanese international university surveyed its students in 2018 and published a study the following year that was approved by several ethical and regulatory boards.

The study found that international students have a higher risk of mental health difficulties than the general population, and that social connectedness (belonging to a social group) and acculturative stress (stress associated with joining a new culture) are predictive of depression.


Explore the `students` data using PostgreSQL to find out if you would come to a similar conclusion for international students and see if the length of stay is a contributing factor.

Here is a data description of the columns you may find helpful.

| Field Name    | Description                                      |
| ------------- | ------------------------------------------------ |
| `inter_dom`     | Types of students (international or domestic)   |
| `japanese_cate` | Japanese language proficiency                    |
| `english_cate`  | English language proficiency                     |
| `academic`      | Current academic level (undergraduate or graduate) |
| `age`           | Current age of student                           |
| `stay`          | Current length of stay in years                  |
| `todep`         | Total score of depression (PHQ-9 test)           |
| `tosc`          | Total score of social connectedness (SCS test)   |
| `toas`          | Total score of acculturative stress (ASISS test) |

```python
-- Run this code to view the data in students
SELECT * 
FROM students;
```

Exploratory Queries to Check
- Do international students face more depression than domestic ones?
- Does length of stay reduce mental stress or depression?
- Does social connectedness protect against depression?
- How do language skills affect mental health?
- Are graduate students more stressed than undergraduates?
- Does gender play a role?
- Is acculturative stress a good predictor of depression?

```python
-- Distribution of Student Types
SELECT inter_dom, COUNT(*) AS student_count
FROM students
WHERE inter_dom IS NOT NULL AND inter_dom != ''
GROUP BY inter_dom;
```

```python
-- Average Depression Score by Student Type
SELECT inter_dom, ROUND(AVG(todep),2) AS avg_depression
FROM students
WHERE inter_dom IS NOT NULL AND inter_dom != ''
GROUP BY inter_dom
```

```python
-- Depression vs Length of Stay
SELECT ROUND(stay) AS stay_year,
		ROUND(AVG(todep),2) AS avg_depression
FROM students
GROUP BY stay
ORDER BY stay_year;
```

```python
-- Average Social Connectedness by Student Type
SELECT inter_dom, ROUND(AVG(tosc),2) AS avg_social_connectedness
FROM students
WHERE inter_dom IS NOT NULL AND inter_dom != ''
GROUP BY inter_dom;
```

```python
-- Average Acculturative Stress by Student Type
SELECT inter_dom, ROUND(AVG(toas),2) AS avg_acculturative_stress
FROM students
WHERE inter_dom IS NOT NULL AND inter_dom != ''
GROUP BY inter_dom;
```

```python
-- Depression by Gender
SELECT gender, ROUND(AVG(todep),2) AS avg_depression
FROM students
	WHERE inter_dom IS NOT NULL AND inter_dom != ''
GROUP BY gender;
```

```python
-- Relationship Between Social Connectedness and Depression
SELECT ROUND(tosc/10.0)*10 AS social_group,
        ROUND(AVG(todep),2) AS avg_depression
FROM students
WHERE tosc IS NOT NULL
GROUP BY social_group
ORDER BY social_group
```

```python
-- Impact of Japanese Language Proficiency on Acculturative Stress
SELECT japanese_cate, ROUND(AVG(toas),2) AS avg_acculturative_stress
FROM students
WHERE japanese_cate != ''
GROUP BY japanese_cate;
```

```python
-- Impact of English Proficiency on Depression
SELECT english_cate, ROUND(AVG(todep),2) AS avg_depression
FROM students
WHERE english_cate IS NOT NULL AND english_cate != ''
GROUP BY english_cate;
```

```python
-- Depression by Academic Level
SELECT academic, ROUND(AVG(todep),2) AS avg_depression
FROM students
	WHERE academic != ''
GROUP BY academic;
```

```python
-- Correlation Between Acculturative Stress and Depression
SELECT ROUND(toas/10.0)*10 AS acc_stress_group,
		ROUND(AVG(todep),2) AS avg_depression
FROM students
WHERE  toas IS NOT NULL
GROUP BY acc_stress_group
ORDER BY acc_stress_group;
```

```python
-- Age and Depression Relationship
SELECT age, ROUND(AVG(todep),2) AS avg_depression
FROM students
WHERE age IS NOT NULL
GROUP BY age
ORDER BY age;
```
